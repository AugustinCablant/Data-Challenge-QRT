Regression

```python
X = df[['rm']].values
y = df['medv'].values

sc_x = StandardScaler()
sc_y = StandardScaler()

X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y[:, np.newaxis]).flatten()

lr = LinearRegression()
lr.fit(X,y_std)

y_pred = lr.predict(X)

print('slope : %.3f' % lr.coef_[0])
print('Intercept : %.3f' % lr.intercept_)
```
